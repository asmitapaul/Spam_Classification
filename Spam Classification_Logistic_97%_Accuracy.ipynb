```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
file_path=r'C:/Users/91861/Downloads/Spam_Dataset/SMSSpamCollection'
```

```python
with open(file_path, 'r', encoding='utf-8') as file:
    corpus = file.readlines()
```

```python
corpus[:10]
```

```python
#Seperating X and Y variables:
X_raw=['' for _ in range(len(corpus))]
Y=['' for _ in range(len(corpus))]
```

```python
for i in range(len(corpus)):
    line=corpus[i]
    Y[i]=line.split("\t")[0]
    if Y[i]=='spam':
        Y[i]="1"
    else:
        Y[i]="0"
    X_raw[i]=line.split("\t")[1]
    X_raw[i]=X_raw[i].replace("\n","")
    X_raw[i]=X_raw[i].lower() #converting everything to lowercase
    
```

```python
X_raw[:10]
```

```python
Y[:10]
```

```python
#Cleaning up the data:
import re
X_raw_cleaned=[]
for i in range(len(X_raw)):
    sent=re.sub('[^a-zA-Z ]','',X_raw[i])
    sent=sent.lower()
    X_raw_cleaned.append(sent)
```

```python
X_raw_cleaned
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
#Tokenizing data, Lemmatizing data and removing stopwords:
para_lemmatized=['' for _ in range(len(X_raw_cleaned))]
for i in range(len(X_raw_cleaned)):
    words=nltk.word_tokenize(X_raw_cleaned[i])
    sent_lemmatized=''
    for word in words:
        if word not in set(stopwords.words('English')):
            sent_lemmatized=sent_lemmatized+" "+lemmatizer.lemmatize(word)
    para_lemmatized[i]=sent_lemmatized
```

```python
para_lemmatized
```

```python
tfidf=TfidfVectorizer()
```

```python
#Applying TF-IDF
X=tfidf.fit_transform(para_lemmatized)
```

```python
#Seperating the dataset into train and test data:
X=pd.DataFrame(X.toarray())# First converting sparse matrix to numpy array and then to dataframe
X_train=X.iloc[:5000,:]
```

```python
X_test=X.iloc[5000:,:]
```

```python
Y_train=Y[:5000]
Y_test=Y[5000:]
```

```python
print(len(Y_train))
print(len(Y_test))
```

```python
#Applying logistic regression model:
Logistic_model=LogisticRegression()
Logistic_model.fit(X_train,Y_train)
```

```python
Y_hat=Logistic_model.predict(X_test)
```

```python
#Calculating model accuracy
accuracy = accuracy_score(Y_test, Y_hat)
```
